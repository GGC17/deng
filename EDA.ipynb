## Let's explore

```python
total_cases_year = pd.DataFrame(deng.groupby(by='year')['total_cases'].sum()).reset_index()
total_cases_sj = pd.DataFrame(deng_sj.groupby(by='year')['total_cases'].sum()).reset_index()
total_cases_iq = pd.DataFrame(deng_iq.groupby(by='year')['total_cases'].sum()).reset_index()
```

```python
fig = plt.figure(figsize=(17, 7))
plt.bar(x=total_cases_year['year'], height=total_cases_year['total_cases'], width=0.9, align='center', label="Total Cases")
plt.bar(x=total_cases_sj['year'], height=total_cases_sj['total_cases'], width=0.9/2, align='center', label="Total Cases San Juan")
plt.bar(x=total_cases_iq['year'], height=total_cases_iq['total_cases'], width=0.9/4, align='center', label="Total Cases Iquitos")
plt.legend()
plt.show()
```

We can see there was a peak in the year of 1994 of dengue cases in SJ, and also a lot of cases in the year of 1998. The data for dengue cases in Iquitos only starts in 2001

##### Dive deeper

```python
deng_city = pd.DataFrame(deng.groupby(['city'])['total_cases'].sum()).reset_index()
plt.bar(deng_city['city'], deng_city['total_cases'])
```

San Juan has the majority of dengue cases in our dataset

```python
years_sj = deng_sj.year.unique()
for year in years_sj:
    plt.plot(deng_sj[deng_sj['year']==year]['weekofyear'], deng_sj[deng_sj['year']==year]['total_cases'])
    plt.title(year)
    plt.show()
```

For San Juan there is usually a peak by the end of the year

```python
years_iq = deng_iq.year.unique()
for year in years_iq:
    plt.plot(deng_iq[deng_iq['year']==year]['weekofyear'], deng_iq[deng_iq['year']==year]['total_cases'])
    plt.title(year)
    plt.show()
```

For Iquitos we can't see any pattern
